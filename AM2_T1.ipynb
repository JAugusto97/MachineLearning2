# Classification

## Instantiating ML Models

## Heterogenic Ensemble of ML Models

````
classifiers = {
    "svm": SVC(probability=True),
    "logistic": LogisticRegression(),
    "dt": DecisionTreeClassifier(),
    "knn": KNeighborsClassifier(),
    "mlp": MLPClassifier(),
    "bayes": GaussianNB(),
}
````

````
class ClassifierEnsemble():
  def __init__(self, classifiers, voting_method):
    self.classifiers = classifiers
    self.voting_method = voting_method
    self.trained = False

  def fit(self, X, y):
    for clf_name, clf in tqdm(self.classifiers.items(), total=len(self.classifiers)):
      clf.fit(X, y)

    if self.voting_method == "weighted":
        self.get_weight(X, y)

    self.trained = True

  def get_weight(self, X, y):
      self.classifier_weights = []
      for clf_name, clf in tqdm(self.classifiers.items(), total=len(self.classifiers)):
        score = statistics.mean(cross_val_score(clf, X, y, cv=5))
        self.classifier_weights.append(score)

  def predict(self, X):
    if not self.trained:
        print("Classifier not trained...")
    
    else:
        if self.voting_method == "hard":
            predictions = []
            for clf_name, clf in self.classifiers.items():
                output = clf.predict(X)
                predictions.append(output)
            predictions = np.array(predictions)

            final_pred = []
            # get the most frequent prediction for each column
            for col in range(predictions.shape[1]):
                unique, counts = np.unique(predictions[:, col], return_counts=True)
                votes = dict(zip(unique, counts)) # construct dict {class: counts}
                ensemble_pred = max(votes, key=votes.get) # get the key for the most frequent value

                final_pred.append(ensemble_pred)

            return final_pred

        elif self.voting_method == "soft":
            predictions = []
            for clf_name, clf in self.classifiers.items():
                output = clf.predict_proba(X) # returns [prob_class0, prob_class1] for each ex
                predictions.append(output)

            predictions = np.array(predictions) # 3d matrix (n_clf, n_sample, n_class)
            votes = {}
            for ex in range(predictions.shape[1]):
                votes[ex] = {}
                for class_type in range(predictions.shape[2]):
                    votes[ex][class_type] = np.sum(predictions[:, ex, class_type])/predictions.shape[0]
                    # {ex1: {class1: avg_prob1, class2: avg_prob2}, ...}
                
            final_pred = []
            for ex, probas in votes.items():
                final_pred.append(max(probas, key=probas.get))

            return final_pred

        elif self.voting_method == "weighted":
            predictions = []
            for clf_name, clf in self.classifiers.items():
                output = clf.predict(X)
                predictions.append(output)
            
            predictions = np.array(predictions)
            final_pred = []
            for col in range(predictions.shape[1]):
                votes =[0]*(predictions.shape[0])
                for row in range(predictions.shape[0]):
                    votes[predictions[row,col]] += predictions[row,col]*self.classifier_weights[row]

                ensemble_pred = max(votes)
                ensemble_pred = votes.index(ensemble_pred)
                final_pred.append(ensemble_pred)

          
            return final_pred
            

        else:
            raise NotImplementedError
````

## Prediction

````
TEST_SIZE = 0.3

X_train_blood, X_test_blood, y_train_blood, y_test_blood = train_test_split(blood_df_t.iloc[:, :-1], blood_df_t.iloc[:, -1], test_size=TEST_SIZE)
X_train_mushroom, X_test_mushroom, y_train_mushroom, y_test_mushroom = train_test_split(mushroom_df.iloc[:, 1:], mushroom_df.iloc[:, 0], test_size=TEST_SIZE)
X_train_phoneme, X_test_phoneme, y_train_phoneme, y_test_phoneme = train_test_split(phoneme_df.iloc[:, :-1], phoneme_df.iloc[:, -1], test_size=TEST_SIZE)
````

````
results = {
    "implemented":
        {
            "blood": {
                "soft": [], "hard": [], "weighted": []
            },
            "mushroom": {
                "soft": [], "hard": [], "weighted": []
            },
            "phoneme": {
                "soft": [], "hard": [], "weighted": []
            },
        },
    "sklearn":
        {
            "blood": {
                "soft": [], "hard": []
            },
            "mushroom": {
                "soft": [], "hard": []
            },
            "phoneme": {
                "soft": [], "hard": []
            },
        },
}
````

````
clf1 = ClassifierEnsemble(classifiers, "soft")
clf2 = ClassifierEnsemble(classifiers, "hard")
clf3 = ClassifierEnsemble(classifiers, "weighted")

# blood
clf1.fit(X_train_blood, y_train_blood)
clf2.fit(X_train_blood, y_train_blood)
clf3.fit(X_train_blood, y_train_blood)
results["implemented"]["blood"]["soft"] = clf1.predict(X_test_blood)
results["implemented"]["blood"]["hard"] = clf2.predict(X_test_blood)
results["implemented"]["blood"]["weighted"] = clf3.predict(X_test_blood)

# mushroom
clf1.fit(X_train_mushroom, y_train_mushroom)
clf2.fit(X_train_mushroom, y_train_mushroom)
clf3.fit(X_train_mushroom, y_train_mushroom)
results["implemented"]["mushroom"]["soft"] = clf1.predict(X_test_mushroom)
results["implemented"]["mushroom"]["hard"] = clf2.predict(X_test_mushroom)
results["implemented"]["mushroom"]["weighted"] = clf3.predict(X_test_mushroom)

# phoneme
clf1.fit(X_train_phoneme, y_train_phoneme)
clf2.fit(X_train_phoneme, y_train_phoneme)
clf3.fit(X_train_phoneme, y_train_phoneme)
results["implemented"]["phoneme"]["soft"] = clf1.predict(X_test_phoneme)
results["implemented"]["phoneme"]["hard"] = clf2.predict(X_test_phoneme)
results["implemented"]["phoneme"]["weighted"] = clf3.predict(X_test_phoneme)
````
